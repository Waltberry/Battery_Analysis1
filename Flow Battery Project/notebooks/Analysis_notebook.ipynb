```python
from galvani import BioLogic
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress # linear regression
from src.file_loader import load_mpr
```

```python
# Load data
mpr_file = load_mpr("CP Test_C01.mpr")
df = pd.DataFrame(mpr_file.data)

# Print columns name
print("Columns:", df.columns)

# Get data count
print("Row count:", df.shape[0])

# Plotting
fig, ax = plt.subplots(8, 1, figsize=(10, 16), sharex=True)

# Function to plot data without best-fit line
def plot_data_only(ax, x, y, label, color):
    ax.plot(x, y, label=label, color=color)
    ax.legend()
    ax.grid(True)

# Plot each feature without best-fit line
plot_data_only(ax[0], df['time/s'], df['control/mA'], 'control/mA', 'b')
plot_data_only(ax[1], df['time/s'], df['Ewe/V'], 'Ewe/V', 'r')
plot_data_only(ax[2], df['time/s'], df['I/mA'], 'I/mA', 'g')
plot_data_only(ax[3], df['time/s'], df['dQ/C'], 'dQ/C', 'm')
plot_data_only(ax[4], df['time/s'], df['(Q-Qo)/C'], '(Q-Qo)/C', 'c')
plot_data_only(ax[5], df['time/s'], df['half cycle'], 'half cycle', 'y')
plot_data_only(ax[6], df['time/s'], df['Q charge/discharge/mA.h'], 'Q charge/discharge/mA.h', 'k')
plot_data_only(ax[7], df['time/s'], df['I Range'], 'I Range', 'orange')

# Adding titles and labels
ax[0].set_title('control/mA vs Time')
ax[0].set_ylabel('Current (mA)')

ax[1].set_title('Ewe/V vs Time')
ax[1].set_ylabel('Potential (V)')

ax[2].set_title('I/mA vs Time')
ax[2].set_ylabel('Current (mA)')

ax[3].set_title('dQ/C vs Time')
ax[3].set_ylabel('Charge (C)')

ax[4].set_title('Charge Difference (Q-Qo)/C vs Time')
ax[4].set_ylabel('Charge (C)')

ax[5].set_title('Half Cycle vs Time')
ax[5].set_ylabel('Half Cycle')

ax[6].set_title('Q charge/discharge/mA.h vs Time')
ax[6].set_ylabel('Charge (mA.h)')

ax[7].set_title('I Range vs Time')
ax[7].set_ylabel('I Range')
ax[7].set_xlabel('Time (s)')

# Adjust layout to prevent overlap
plt.tight_layout()

plt.show()
```
