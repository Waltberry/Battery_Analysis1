```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from src.file_loader import load_txt
import seaborn as sns
```

```python
# Load data
df = load_txt('pulseOutput')

print(df.head())
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df["Time Step"], df["voltage_vp"], color='blue', label='Voltage')
plt.title('Scatter Plot of Voltage vs Time Step')
plt.xlabel('Time Step')
plt.ylabel('Voltage (V)')
plt.legend()
plt.grid(True)
plt.show()

# Line Plotsvsvs
plt.figure(figsize=(10, 6))
plt.plot(df["Time Step"], df["voltage_vp"], color='red', label='Voltage', linestyle='-', marker='o')
plt.title('Line Plot of Voltage vs Time Step')
plt.xlabel('Time Step')
plt.ylabel('Voltage (V)')
plt.legend()
plt.grid(True)
plt.show()
```
